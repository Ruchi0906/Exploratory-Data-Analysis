# Replace Null Values

```python
A.isna().sum()
```

```python
def Replace_nan(df):
    q=pd.DataFrame(df.isna().sum(),columns=["Count_nan"])
    for i in q[q.Count_nan>0].index:
        if(df[i].dtypes=='object'):
            x=df[i].mode()[0]
            df[i]=df[i].fillna(x)
        else:
            y=round(df[i].mean(),2)
            df[i]=df[i].fillna(y)
```

```python
Replace_nan(A)
```
